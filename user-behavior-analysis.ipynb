```python
new_column_order = ['Order ID', 'User ID', 'User Name', 'Order Date', 'Meal Type', 'Dish Name',
       'Order Status', 'Amount (USD)', 'Time of Day', 'Rating', 'Session ID',
       'Session Start', 'Session End', 'Duration (mins)', 'Session Rating',
       'Age', 'Location', 'Registration Date', 'Phone', 'Email',
       'Favorite Meal', 'Total Orders']

combined_data = combined_data[new_column_order]
combined_data.head()
```

```python
combined_data.replace('NaN', np.nan, inplace=True)

print(combined_data.isnull().sum())
```

```python
session_count = combined_data.groupby('User ID')['Session ID'].nunique().reset_index()
session_count.rename(columns={'Session ID': 'Total Sessions'}, inplace=True)

order_summary = combined_data.groupby('User ID').agg({
    'Total Orders': 'sum',
    'Amount (USD)': 'sum'}).reset_index()

user_data = pd.merge(order_summary, session_count, on='User ID', how='inner')
user_data.head()
```

```python
correlation_matrix = user_data[['Total Orders', 'Amount (USD)', 'Total Sessions']].corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap = 'Spectral')
plt.title('Correlation Matrix of Total Orders, Amount (USD), and Total Sessions')
plt.show()
```

```python
most_popular = combined_data['Dish Name'].value_counts()
most_popular
```

```python
demographic_analysis = combined_data.groupby(['Age', 'Location']).agg({'Total Orders': 'sum', 'Amount (USD)': 'sum'})
demographic_analysis
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data = user_data, x = 'Total Sessions', y = 'Total Orders')
plt.xticks([1, 2, 3])  
plt.title('Cooking Sessions vs Total Orders')
plt.xlabel('Number of Sessions')
plt.ylabel('Total Orders')
plt.show()
```

```python
plt.figure(figsize=(10,5))
most_popular.plot(kind = 'bar', color = 'skyblue')
plt.title('Top Most Popular Dishes')
plt.ylabel('Order Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.pointplot(x='Age', y='Total Orders', hue='Location', data =  demographic_analysis)
plt.title('Pointplot: Total Orders by Age and Location')
plt.xlabel('Age')
plt.ylabel('Total Orders')
plt.show()
```

#       **Finding**


#### 1. Total Orders by Location:
    . Users from New York, San Francisco, and Los Angeles placed the highest number of orders, with New York leading at 36 orders.
    
    . Users from Austin and Miami placed significantly fewer orders (7 and 5 orders, respectively).


#### 2. Spending Patterns:
    . Total spending correlates with the number of orders. For example:
      
        - New York: 36 orders, $35.00 spent.
            
        - San Francisco: 20 orders, $21.50 spent.
        
    . However, Chicago had 15 orders but spent only $8.50, indicating smaller order sizes or lower-value items.

#### 3. Age and Engagement:

    . Younger users (aged 25-28) are more engaged, with higher order counts and spending compared to older users (aged 31+).

    


#### 4. Demographic Factors:

     . Certain locations, like New York and San Francisco, exhibit higher engagement regardless of age, suggesting strong market potential.
     
     . Locations like Miami and Chicago underperform in both order count and spending.
